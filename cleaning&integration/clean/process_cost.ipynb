This file exist because the html of laptopmedia.com change mid-scraping, which require us to re-scrape cost 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = '../data/ignore/'
COST_FILE = 'laptop_2_6_hehe (2).csv'
OLD_COST_FILE = 'intermediate.csv'
```

```python
lap = pd.read_csv(data_path + 'Laptop.csv')
```

```python
link = lap['link']
```

```python
cost = pd.read_csv(data_path + COST_FILE)
cost.set_index('link', inplace=True)
```

```python
if OLD_COST_FILE:
    old_cost = pd.read_csv(data_path + OLD_COST_FILE)
else:
    old_cost = lap[['link', 'Cost']].copy()

old_cost.set_index('link', inplace=True)
preupdate_links = old_cost['Cost'].notnull().apply(int)
```

```python
cost_file_unique_links = set(cost.index)
old_cost_unique_links = set(old_cost.index)
print('total updated should be near:', len(cost_file_unique_links.intersection(old_cost_unique_links)))
```

```python
n = 0
for link in cost_file_unique_links:
    n+=1
    print(cost.loc[link])
    old_cost.loc[link] = cost.loc[link].values[0]
    print(n)
```

```python
updated_links = old_cost['Cost'].notnull().apply(int)
print('total updated:', sum(updated_links) - sum(preupdate_links))
```

```python
pd_out = old_cost.copy()
pd_out.to_csv(data_path + 'intermediate.csv', index=True)
```
