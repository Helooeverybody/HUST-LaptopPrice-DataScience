```python
import pandas as pd
import numpy as np
```

```python
data_path = '../data/ignore/'
lap = 'laptop_final.csv'
```

```python
lap = pd.read_csv(data_path + lap)
```

```python
not_lap = ['apple', 'steam']
for brand in not_lap:
    print('Found laptop for brand:', brand)
    for name in lap[lap['name'].str.lower().str.find(brand) != -1]['name'].values:
        print(name)
    lap = lap[lap['name'].str.lower().str.find(brand) == -1]
    print('Removed laptop for brand:', brand, '\n', '-'*30)
```

```python
lap.to_csv(data_path + 'laptop_final_no_apple.csv', index=False)
```
