```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common import TimeoutException
import urllib3
import time
import pandas as pd
import json
import re
```

```python
mine = pd.DataFrame()

hehe = pd.read_csv("better_gpu_link.csv")
```

```python
ffx = webdriver.Chrome()
```

```python
mineeee = list(hehe.iterrows())
counter = 0
```

```python
while True:
    id,row=mineeee[counter]
    print(id)
    link = "https://www.techpowerup.com" + row["link"]
    while True:
        try:
            ffx.get(link)
            
            # he = urllib3.request("GET",link)
            # soup = BeautifulSoup(he.data)
        
            soup = BeautifulSoup(ffx.page_source)
            better = soup.find("div", attrs = {'class':"sectioncontainer"})
            hoho = better.find_all("dl")
        except AttributeError:
            ffx = webdriver.Chrome()
        else:
            break
    for ho in hoho:
        try:
            a = ho.find("dt").contents[0]
        except AttributeError:
            continue
        if a[-1] == ":":
            a = a[:-1]
        if a not in ["Architecture","GPU Name","Generation","Predecessor","Successor","Reviews"]:
            try:
                b = ho.find("dd").contents[0].strip()
            except IndexError:
                b = None
        else:
            b = ho.find("dd").find("a").contents[0].strip()
        mine.at[id,a] = b
    # break
    counter+=1
```
