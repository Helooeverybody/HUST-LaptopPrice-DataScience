```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data_path = '../data/'
gpu_data = 'gpu.csv'
pd1 = pd.read_csv(data_path + gpu_data, dtype=str)
pd1['name'] = pd1['name'].str.lower()
```

```python
# Ranking: extract the number
rank = pd1['Ranking'].str.extract(r'(\d+)')
rank.rename(columns={0: 'Ranking'}, inplace=True)
rank
```

```python
# Manufactoring process: extract the number
manuf = pd1['Manufacturing Process'].str.extract(r'(\d+)')
manuf.rename(columns={0: 'Manufacturing process'}, inplace=True)
manuf
```

```python
# Architecture: ohe later, leave as is for now
arch = pd1['Architecture']
arch
```

```python
# Base frequency: extract the number
base_freq = pd1['Base frequency'].str.extract(r'(\d+)')
base_freq.rename(columns={0: 'Base Frequency'}, inplace=True)
base_freq
```

```python
# Memory type: ohe later, leave as is for now
mem_type = pd1['Memory Type']
mem_type
```

```python
# Memory frequency: extract the number
mem_freq = pd1['Memory Frequency'].str.extract(r'(\d+)')
mem_freq.rename(columns={0: 'Memory Frequency'}, inplace=True)
mem_freq
```

```python
# Memory bandwidth: extract the number
mem_bw = pd1['Memory bandwidth'].str.extract(r'(\d+)')
mem_bw.rename(columns={0: 'Memory Bandwidth'}, inplace=True)
mem_bw
```

```python
# Release quarter: extract number as convention: Q1 2024 = 2024.00, Q2 2024 = 2024.25, ...
release = pd1['Released'].str.extract(r'(\d+)\s(\d+)').astype(float)
release = release[1] + (release[0] - 1) / 4
release = pd.DataFrame(release)
release.rename(columns={0: 'Released'}, inplace=True)
release
```

```python
# Power consumption: extract the number
power = pd1['Power consumption'].str.extract(r'(\d+)').astype(float)
power.rename(columns={0: 'Power Consumption'}, inplace=True)
power
```

```python
# Cores: extract the number
cores = pd1['Cores'].str.extract(r'(\d+)')
cores.rename(columns={0: 'Cores'}, inplace=True)
cores
```

```python
# Max frequency: extract the number
max_freq = pd1['Maximum frequency'].str.extract(r'(\d+)')
max_freq.rename(columns={0: 'Max Frequency'}, inplace=True)
max_freq
```

```python
# Memory capacity: extract the number
mem_cap = pd1['Memory Capacity'].str.extract(r'(\d+)')
mem_cap.rename(columns={0: 'Memory Capacity'}, inplace=True)
mem_cap
```

```python
# Memory bus: extract the number
mem_bus = pd1['Memory bus'].str.extract(r'(\d+)')
mem_bus.rename(columns={0: 'Memory Bus'}, inplace=True)
mem_bus
```

```python
# Drivers: get the name of the websute
drivers = pd1['Drivers'].str.extract(r'\.(\w+)\.')
drivers.rename(columns={0: 'Drivers'}, inplace=True)
drivers
```

```python
# AI cores: extract the number
ai_cores = pd1['AI Cores'].str.extract(r'(\d+)')
ai_cores.rename(columns={0: 'AI Cores'}, inplace=True)
ai_cores
```

```python
# AI performance: extract the number
ai_perf = pd1['AI Performance (INT8)'].str.extract(r'(\d+)')
ai_perf.rename(columns={0: 'AI Performance'}, inplace=True)
ai_perf
```

```python
essentials = pd1[['gpu_link', 'name']]
cols = [essentials, rank, manuf, arch, base_freq, mem_type, mem_freq, mem_bw, release, power, cores, max_freq, mem_cap, mem_bus, drivers, ai_cores, ai_perf]

pd_out = pd.concat(cols, axis=1)
pd_out
```

```python
pd_out.to_csv(data_path + 'gpu_cleaned.csv', index=False)
```
